##Data visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import numpy as np
```

```python
sns.heatmap(df.corr(),annot=True);#annot=True = grafiğin içindeki korelasyon kutucuklarına korelasyon değerlerini yazdırır
#korelasyon %100 çıkarsa veri farklı şekilde kendini tekrar etmiş olur.
```

```python
plt.figure(figsize=(8,6)) #grafiğin boyutunu ayarlar
sns.countplot(df['Position']); #hangi aylarda satış yapılmış gösterir
#countplot=sayıları sayar sonucu gösterir
plt.xticks(rotation=45); #altta ki yazıları 45 derece ile yazmak için
```

```python
plt.figure(figsize=(8,6)) #grafiğin boyutunu ayarlar
sns.countplot(df['Breed']); #hangi aylarda satış yapılmış gösterir
#countplot=sayıları sayar sonucu gösterir
plt.xticks(rotation=90); #altta ki yazıları 45 derece ile yazmak için
```

```python
plt.figure(figsize=(16,8)) #grafiğin boyutunu ayarlar
sns.countplot(df['Registrations']); #hangi aylarda satış yapılmış gösterir
#countplot=sayıları sayar sonucu gösterir
plt.xticks(rotation=90); #altta ki yazıları 45 derece ile yazmak için
```

```python
plt.figure(figsize=(16,8)) #grafiğin boyutunu ayarlar
sns.countplot(df['Year']); #hangi aylarda satış yapılmış gösterir
#countplot=sayıları sayar sonucu gösterir
plt.xticks(rotation=90); #altta ki yazıları 45 derece ile yazmak için
```

```python
plt.figure(figsize=(15,15))
explode=[0.1,0,0] #en büyük dilimi ayırmak için
df['Breed'].value_counts().plot.pie(autopct='%1.1f%%') #aylara göre satışı pasta grafiğinde gösterir
#yüzdeleri dilimlerin içine yazdırmak için(autopct='%1.1f%%')
plt.title('Köpek cinsleri')
```

```python
plt.figure(figsize=(12,6))
sns.countplot(df['Year'],hue=df.Breed)#hangi ürünler en çok satılmış gösterir#hue=df.Genderhangirenk hangi cinsiyet olduğunu gösterir.
plt.xticks(rotation=90);
#hue seçilen sütunu renklendirmek için
```

```python
plt.figure(figsize=(12,6))
sns.countplot(df['Breed'],hue=df['Position'])
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df['Breed'],y=df.index,palette='hot')
plt.xticks(rotation=90);
```

```python
sns.boxplot(x=df['Year'])
```

```python
plt.figure(figsize=(10,8))
sns.barplot(y=df['Position'],x=df['Breed']);#hangi üründen ne kadar gelir elde etmişiz
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x="Breed",y="Year",data=df)
plt.xticks(rotation=90);
```
